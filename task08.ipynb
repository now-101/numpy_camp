#### Figure和subplot

matplotlib的图像都位于Figure对象中。你可以用plt.figure创建一个新的Figure：

```python
fig = plt.figure()
```

如果⽤的是IPython，这时会弹出⼀个空窗⼝，但在Jupyter中，必须再输⼊更多命令才能看到。 plt.figure有⼀些选项，特别是figsize，它⽤于确保当图⽚保存到磁盘时具有⼀定的⼤⼩和纵横⽐。

不能通过空Figure绘图。必须⽤add_subplot创建⼀个或多个subplot才⾏：

```python
ax1 = fig.add_subplot(2, 2, 1)
```

这条代码的意思是：图像应该是2×2的（即最多4张图），且当前选中的是4个subplot中的第⼀个（编号从1开始）。如果再把后⾯两个subplot也创建出来，最终得到的图像如图9-2所示

```python
ax2 = fig.add_subplot(2,2,2)
```

```python
ax3 = fig.add_subplot(2,2,3)
```

```python
fig2 = plt.figure()
ax12 = fig2.add_subplot(2,2,1)
ax22 = fig2.add_subplot(2,2,2)
ax32 = fig2.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), 'k--')
_ = ax12.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax22.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

"k--"是⼀个线型选项，⽤于告诉matplotlib绘制⿊⾊虚线图。上⾯
那些由fig.add_subplot所返回的对象是AxesSubplot对象，直接
调⽤它们的实例⽅法就可以在其它空着的格⼦⾥⾯画图了，如图
9-4所示：

```python
_ = ax12.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax22.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```
